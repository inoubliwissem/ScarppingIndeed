```python
import csv
from datetime import datetime
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#create an url based on title and location
def get_url(position, location):
    """Generate url from position and location"""
    template = 'https://www.indeed.com/jobs?q={}&l={}'
    position = position.replace(' ', '+')
    location = location.replace(' ', '+')
    url = template.format(position, location)
    return url
```

```python
#get full description of a Job (url)
def getFullDescription(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    jobTitle = soup.find_all('div', class_='jobsearch-JobInfoHeader-title-container') 
    jobDscription = soup.find_all('div', class_='jobsearch-jobDescriptionText')
    result={}
    result['title']=jobTitle[0].text
    result['Description']=jobDscription[0].text
    return result
```

```python
#url as an example
# get some JobOffers and store them into a dataframe
titles=[]
descriptions=[]
for p in range(2,12):
    url="https://uk.indeed.com/jobs?q=SALES&l=United%20Kingdom&start="+str((p*10))
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    frames=soup.find_all('div', 'mosaic-provider-jobcards')
    divcard=frames[0]
    for c in divcard.children:
        blockType=c.name
        if blockType=="a":
            link=c.get("href")
            urlJob="https://uk.indeed.com"+link
            results=getFullDescription(urlJob)
            titles.append(results['title'])
            descriptions.append(results['Description'])
            print (urlJob)
rows = {'Title': titles, 'Description': descriptions}
df = pd.DataFrame(data=rows)
df.to_csv("SALES.csv")
```
